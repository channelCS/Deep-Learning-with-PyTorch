### Understanding Tensors through Numpy

```python
# 0-D
x0 = 3

# 1-D
x1 = np.array([10,20,30])

# 2-D
x2 = np.zeros((3,3))

# 3-D
x3 = np.ones((3,3,4))

# 4-D
x4 = np.random.randn(4,3,3,4)
```

```python
print(x1.shape, x2.shape, x3.shape, x4.shape)
```

### Initializing PyTorch Tensors

### Understanding Tensors through PyTorch

```python
# 0-D
x0 = 3

# 1-D
x1 = torch.Tensor([10,20,30])

# 2-D
x2 = torch.zeros((3,3))

# 3-D
x3 = torch.ones((3,3,4))

# 4-D
x4 = torch.randn(4,3,3,4)
```

```python
print(x1.shape, x2.shape, x3.shape, x4.shape)
```
